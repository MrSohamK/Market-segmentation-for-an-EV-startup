```python
# encoding the categorical features

# PowerTrain feature
df['PowerTrain'].replace(to_replace=['RWD','FWD','AWD'],value=[0, 1, 2],inplace=True)

# RapidCharge feature
df['RapidCharge'].replace(to_replace=['No','Yes'],value=[0, 1],inplace=True)
```

```python
# selecting features for building a model
X = df[['AccelSec','TopSpeed_KmH','Efficiency_WhKm','FastCharge_KmH', 'Range_Km', 'RapidCharge', 'Seats', 'PriceEuro','PowerTrain']]
```

```python
# feature scaling
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# applying Principle Component Analysis (PCA)
pca = PCA(n_components=9)
X_pca = pca.fit_transform(X_scaled)
df_pca = pd.DataFrame(X_pca, columns=['PC1', 'PC2', 'PC3', 'PC4', 'PC5', 'PC6', 'PC7', 'PC8', 'PC9'])
df_pca.head()
```

```python
# plotting the results of Elbow

wcss = []

for i in range(1, 11):
  kmean = KMeans(n_clusters=i, init='k-means++', random_state=90)
  kmean.fit(X_pca)
  wcss.append(kmean.inertia_)

plt.figure(figsize=(8,6))
plt.title('Plot of the Elbow Method', size=15, family='serif')
plt.plot(range(1, 11), wcss)
plt.xticks(range(1, 11), family='serif')
plt.yticks(family='serif')
plt.xlabel('Number of Custers (K)', family='serif')
plt.ylabel('WCSS', family='serif')
plt.grid()
plt.tick_params(axis='both', direction='inout', length=6, color='purple', grid_color='lightgray', grid_linestyle='--')
plt.show()
```

```python
# training the model using k=4 as rendered by the above plot
kmean = KMeans(n_clusters=4, init='k-means++', random_state=90)
kmean.fit(X_pca)
```

```python

# check the labels assigned to each data point
print(kmean.labels_)
```

```python

# check the size of clusters
pd.Series(kmean.labels_).value_counts()
```

```python

# adding a new feature of cluster labels to the dataset 3
df['clusters'] = kmean.labels_
```

```python
# visualizing clusters
plt.figure(figsize=(7,5))

sns.scatterplot(data=df_pca, x='PC1', y='PC9', s=70, hue=kmean.labels_, palette='coolwarm', zorder=2, alpha=.9)
plt.scatter(x=kmean.cluster_centers_[:,0], y=kmean.cluster_centers_[:,1], marker="*", c="magenta", s=80, label="Centroids")

plt.xlabel('PC1', family='serif', size=12)
plt.ylabel('PC9', family='serif', size=12)
plt.xticks(family='serif')
plt.yticks(family='serif')
plt.grid()
plt.tick_params(grid_color='lightgray', grid_linestyle='--', zorder=1)
plt.legend(title='Labels', fancybox=True, shadow=True)
plt.title('K-Means Clustering Results', family='serif', size=15)
plt.show()
```

```python
numericalColumns = finalDF.select_dtypes(include=['int64', 'float64']).columns
datetimeColumns = finalDF['Month']
numericalData = finalDF[numericalColumns]
```

```python
dataPCA = pca.transform(scaledData)
```

```python

dfPCA = pd.DataFrame(dataPCA, columns = [f'PC{x +1}' for x in range(len(scaledData.columns))])
```

```python
dfPCA.head()
```

```python
PCAsummary = pd.DataFrame({'Standard Deviation':dfPCA.std(),
                           'Proportion of Variance': pca.explained_variance_ratio_,
                           'Cumultive Proportion': pca.explained_variance_ratio_.cumsum()})
```

```python
PCAsummary
```

```python
factorLoadings = pd.DataFrame(pca.components_, columns = scaledData.columns, index = dfPCA.columns).T
```

```python
segment = electricBOV.merge(hybridCombined, on = 'Cluster Number', how = 'left')
segment
```

```python
# Target segments

plt.figure(figsize = (12,4))
sns.scatterplot(x = "ELECTRIC(BOV)", y = "HYBRID COMBINED",data=segment,s=400, hue='Cluster Number', palette='coolwarm')
plt.title("Simple segment evaluation plot for the Electrical data set",
          fontsize = 17)
plt.xlabel("Electric (BOV) Vehicles", fontsize = 14)
plt.ylabel("(Petrol + Diesel) Hybrid Vehicles", fontsize = 14)
plt.legend(title='Segment', loc='lower right')
plt.show()
```

```python
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.impute import SimpleImputer


features = ['ELECTRIC(BOV)', 'HYBRID COMBINED']
target = 'Cluster Number'

X = segment[features]
y = segment[target]

# Creating a decision tree
tree = DecisionTreeClassifier()
tree.fit(X, y)

# Visualize the decision tree
plt.figure(figsize=(12, 8))
plot_tree(tree, feature_names=features, class_names=[str(i) for i in segment[target].unique()], filled=True, rounded=True)
plt.show()
```
